```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

web = webdriver.Firefox()
web.get('https://kkutu.world/?server=0')
usr_name = web.find_elements(By.CLASS_NAME, 'my-stat-name')

user_id = usr_name[0].text
print("welcome .. " + user_id)

f = open('kr_korean.csv', 'r', encoding='utf-8-sig')
reader = csv.reader(f)

result = []
for line in reader:
    result.append(line)

already_use = []

print('데이터 개수: ' + str(len(result)) + '개')
f.close()
```

```python
def get_now_word():
    """_summary_

    Returns:
        word (str) : 현재 글자를 받아옵니다.
    """
    word = web.find_element(By.CSS_SELECTOR, '.jjo-display')
    return word.text
```

```python
def send_word(text:str):
    """text으로 입력받은 글자를 보냅니다.

    Args:
        text (str): 보낼글자.
    """
    
    i = web.find_element(By.ID, 'Talk')
    i.send_keys(text)
    web.find_element(By.ID, 'ChatBtn').click()
    
    print("Sended Word : " + text)
```

```python
def get_word(a : str):
    """a 라는 단어로 시작하는 문장을 가져옵니다.

    Args:
        a (str): 첫글자 ex. (가, 이, 고)

    Returns:
        list : a라는 단어로 시작하는 문장들의 리스트
    """
    
    list_of_text = []
    for i in result:
        if i[0][0] == a and not i[0] in already_use:
            if "-" in i[0] or "^" in i[0] or "*" in i[0] or " " in i[0] or "" in i[0] or "ㆍ" in i[0]:
                continue
            text = i[0]
            if len(text) >= 2:
                list_of_text.append(text)
    list_of_text.sort(key=lambda item: (len(item), item), reverse=True)
    return list_of_text
```

```python
import time
import re

def attack(start):
    """받은 첫단어로 그의 맞는 문장을 찾아 공격합니다.

    Args:
        start (_type_): 첫단어
    """
    lst = get_word(start)
    now_word = get_now_word()
                
    if '(' in now_word:
        now_word = now_word.split('(')[0]

    if len(lst) == 0:
        send_word(start + " ㅅ12ㅂ ")
        print('사전이 없습니다.')
        return False
    
    send_word(lst[0])
    already_use.append(lst[0])
    return True
```

```python
import csv

def reload_csv():
    global result
    
    f = open('kr_korean.csv', 'r', encoding='utf-8-sig')
    reader = csv.reader(f)

    result = []
    for line in reader:
        result.append(line)
```

```python
import csv

def add_words(words):
    """
    특정 글자를 csv에 추가합니다\n
    혹은 현제 사용된 언어에 추가합니다.

    Args:
        words (list): 글자를 받아옵니다.
    """
    
    adding_word = True
    
    for i in words:
        if "/" in i or "·" in i or i == "명사" or i == "형용사":
            pass
        else:
            f = open('kr_korean.csv', 'a', encoding='utf-8-sig', newline='')
            c = csv.writer(f)
            
            for e in result:
                if e[0] == i:
                    adding_word = False
            
            print(adding_word)
            
            if adding_word == True:
                c.writerow([i, '명사'])
                f.close()
```

```python
import time

cant_found = False
get_history = False
check_death = False

def auto():
    while True:
        if get_now_word() == "게임 끝!":
            send_word("GGS!")
            break
        
        try:
            
            global check_death
            global get_history
            global cant_found
            
            try:
                if check_death == False:
                    me = WebDriverWait(web, 0).until(EC.presence_of_element_located((By.CLASS_NAME, 'game-user-bomb')))
                    already_use.clear()
                    reload_csv()
                    print("Resetting Array...")
                    cant_found = False
                    check_death = True
            except:
                pass
            
            t = web.find_elements(By.CLASS_NAME, 'game-user-current')
            n = t[0].find_elements(By.CLASS_NAME, 'game-user-name')
            plr_name = n[0].text
            
            if str(plr_name) == str(user_id):
                now_word = get_now_word()
                
                if '(' in now_word:
                    now_word = now_word.split('(')
                
                if len(now_word) == 1:
                    if cant_found != True:
                        x = attack(now_word)
                        if x == False:
                            cant_found = True
                    else:
                        continue
                elif type(now_word) == list:
                    if cant_found != True:
                        a = True
                        for i in range(len(now_word)):
                            wordv = get_word(now_word[i][:1])
                            if len(wordv) != 0:
                                print("Selected Word : " + wordv[0])
                                a = False
                                break
                        if a == False:
                            already_use.append(wordv[0])
                            send_word(wordv[0])
                        else:
                            cant_found = True
                    else:
                        continue
                
                get_history = False
                check_death = False
                
            else:
                try:
                    if get_history == False:
                        k = web.find_elements(By.CLASS_NAME, 'history')
                        add_words(k[0])
                        get_history = True
                    else:
                        pass
                except:
                    pass
            
        except:
            continue

auto()
```
